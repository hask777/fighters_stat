```python
import json
import pandas as pd

file = "fighter_name.json"
data =None
with open(file, 'r') as f:
    data = json.load(f)
    
# print(data)
```

```python
def get_fighter_knocks(player):   
    knockOuts = player['TechnicalKnockouts']

    if type(knockOuts) != int:
        knockOuts = 1

    return knockOuts

fighters = sorted(data, key=get_fighter_knocks, reverse=True)

stats = []
for fighter in fighters:
    f_name = fighter["FirstName"] + " " + fighter["LastName"]
    wieightclass = fighter["WeightClass"]
    wins = fighter["Wins"]
    losses = fighter["Losses"]
    draws = fighter["Draws"]
    knockoutswins = fighter["TechnicalKnockouts"]
    knockoutslosses = fighter["TechnicalKnockoutLosses"]
    submissions = fighter["Submissions"]
    sublosses = fighter["SubmissionLosses"]
    
    stats.append([f_name, wieightclass, wins, losses, draws, knockoutswins, knockoutslosses, submissions, sublosses])
```

```python
df = pd.DataFrame(stats, columns=["title", "wieightclass", "wins", "losses", "draws", "knockoutswins", "knockoutslosses", "submissions", "sublosses"])
df.head(20)
```

```python
top10 = df.head(10)
ax = top10.plot.bar(x="title", y="knockoutswins", figsize=(12,8), fontsize=14)
```
